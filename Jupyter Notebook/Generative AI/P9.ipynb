```python
Question 5: Code Generation
Scenario:
You are learning Java programming.
Use BRAD to generate a Java program to check whether a number is prime, with short explanation comments.
Solution:
B - Background: You are practicing Java programming basics.
R - Role: Act as a Java tutor.
A - Action: Generate a Java program to check prime numbers.
D - Details: Include short explanation comments.
Final Prompt (BRAD):
You are learning Java programming.
Act as a Java tutor.
Generate a Java program that checks whether a number is prime.
Include short, beginner-friendly explanation comments in the code.
Expected Output / Analysis:
Output Example:
public class PrimeCheck {
    public static void main(String[] args) {
        int n = 17; // number to check
        boolean isPrime = true;

        // check divisors from 2 to n/2
        for (int i = 2; i <= n / 2; i++) {
            if (n % i == 0) {
                isPrime = false; // not prime
                break;
            }
        }

        if (isPrime) System.out.println(n + " is prime.");
        else System.out.println(n + " is not prime.");
    }
}
Analysis:
The code is simple, includes comments, and fulfills the prime-checking requirement.
```
