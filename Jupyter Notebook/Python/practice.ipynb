```python
n = int(input("Enter a number : "))
flag = 1
for i in range(2,int(n**0.5)):
    if(n % i == 0):
        flag = 0
if(flag):
    print(n," is a prime number")
else:
    print(n," is not a prime number")
```

```python
n = int(input("Enter the limit : "))
for i in range(2,n+1):
    flag = 1
    j = 2
    while(j*j <= i):
        if(i % j == 0):
            flag = 0
        j+=1
    if(flag):
        print(i, end = " ")
```

```python
m, n = map(int, input("Enter the start and end limit: ").split())
for i in range(m,n+1):
    flag = 1
    j = 2
    while(j*j <= i):
        if(i % j == 0):
            flag = 0
        j+=1
    if(flag):
        print(i, end = " ")
```
